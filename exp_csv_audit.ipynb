```python
import pandas as pd
import numpy as np
import os

def auditar_datasets(rutas_csv):
    for ruta in rutas_csv:
        print(f"\n{'='*55}")
        print(f"🔍 Auditando: {ruta}")
        print(f"{'='*55}")

        # 1. Convertir CSV a DataFrame
        if not os.path.exists(ruta):
            print(f"❌ El archivo {ruta} no existe. Verifica la ruta.")
            continue
            
        # Así se carga el CSV a un DataFrame
        df = pd.read_csv(ruta)

        # 2. Forma del dataset
        print(f"\n📊 1. Forma del Dataset:")
        print(f"  - Imágenes procesadas (Filas): {df.shape[0]}")
        print(f"  - Características + Etiqueta (Columnas): {df.shape[1]}")

        # 3. Balance de clases
        print(f"\n⚖️ 2. Balance de Clases (Columna 'clase'):")
        if 'clase' in df.columns:
            conteo = df['clase'].value_counts()
            # Asumiendo que 1 es Neumonía y 0 es Normal según el script anterior
            print(f"  - Clase 1 (Neumonía): {conteo.get(1, 0)}")
            print(f"  - Clase 0 (Normal): {conteo.get(0, 0)}")
        else:
            print("  - ⚠️ No se encontró la columna 'clase'.")

        # 4. Valores nulos o infinitos
        print(f"\n⚠️ 3. Integridad de los Datos:")
        nulos = df.isnull().sum().sum()
        # Verificamos infinitos solo en columnas numéricas
        infinitos = np.isinf(df.select_dtypes(include=np.number)).sum().sum()
        
        print(f"  - Valores nulos (NaN): {nulos} " + ("✅" if nulos == 0 else "❌"))
        print(f"  - Valores infinitos (inf): {infinitos} " + ("✅" if infinitos == 0 else "❌"))

        # 5. Sanity Check de los umbrales de Multi-Otsu
        print(f"\n📏 4. Rango de Umbrales Multi-Otsu (0 a 255):")
        umbrales = ['P_0', 'P_1', 'P_2']
        for u in umbrales:
            if u in df.columns:
                min_val = df[u].min()
                max_val = df[u].max()
                estado = "✅" if (min_val >= 0 and max_val <= 255) else "❌ REVISAR"
                print(f"  - {u}: Min = {min_val:.1f}, Max = {max_val:.1f} {estado}")

# Lista de los 4 CSVs generados por tu pipeline de MicroDE
archivos = [
    "radiomics_DE_rand_1.csv",
    "radiomics_DE_best_1.csv",
    "radiomics_DE_rand_2.csv",
    "radiomics_DE_best_2.csv"
]

# Ejecutar la auditoría
auditar_datasets(archivos)
```
